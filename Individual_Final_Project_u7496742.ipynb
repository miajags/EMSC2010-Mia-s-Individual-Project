# Regression Analysis

Considering the Bird Index and the Average Fairy Wren Population at the ANBG in Canberra, let's see if the FW trend fits with the national BI trend.
Using regression.
Calculate 95% HDI for FWs, and theoretically the BI line should be within/overlap with this confidence interval. But, this relies on the uncertainty of the FW data.

````python
import bambi as bmb #for automated Bayesian regression
import arviz as az #for analysis of Bayesian models
````

````python
#plotting FW data as a scatter plot, year and average population

plt.scatter(year_fw_filtered, average_pop_filtered)
plt.xlabel('Year')
plt.ylabel('Population - Average Number of Breeding FWs')
plt.title('Average Fairy-Wren ANBG Population Over Time')
plt.grid(True)
plt.grid(True, which='minor', linestyle='-', linewidth='0.05', color='black')
plt.minorticks_on()
plt.show()
````

````python
#finding a polynomial to fit the equation

rdata = pd.DataFrame({"x": year_fw_filtered, "y": average_pop_filtered}) #put the data into a dataframe with variable names "x" and "y".
model = bmb.Model("y ~ x", rdata) #setup the first-order polynomial model
# ~ means y can be predicted by x (y function of x)

rdata["x"] = rdata["x"] #used for linear regression
rdata["x2"] = rdata["x"] ** 2 #used for quadratic regression
rdata["x3"] = rdata["x"] ** 3 #used for cubic regression
rdata["x4"] = rdata["x"] ** 4 #used for quartic regression
````

````python
log_pop_data = np.log(average_pop_filtered)
plt.scatter(year_fw_filtered, log_pop_data)
````

````python
#log linear
popdata = pd.DataFrame({"x": year_fw_filtered, "y": log_pop_data}) #put the data into a dataframe with variable names "x" and "y".
model = bmb.Model("y ~ x", popdata) #setup the first-order polynomial model
# ~ means y can be predicted by x (y function of x)

idata = model.fit(draws=2000, tune=2000, chains=8) #sample the posteriors
````

````python
## plot priors and prosterior

model.plot_priors()
plt.tight_layout()
plt.show()

# plot posterior distributions
# this is our visual check that the samples look 'good' - squiggly means the samples did their job

az.plot_trace(idata);
plt.tight_layout() #this spaces the plots out so that they don't overlap

# intercept = intercept of x with y, sigma = standard deviation, x = gradient
````

````python
## predictive 95% HDI

xrange = np.linspace(year_fw_filtered.min(), year_fw_filtered.max(), 16000) # Predict across a range of x values from 1 to 11
new_data = pd.DataFrame({"x": xrange}) #dataframe with the new x-values

# Ensure both response_params (for hdi_mean) and response (for y_pps_draws) are predicted
model.predict(idata, data=new_data, kind='response_params') #predict the distribution of regression lines at each x-value
model.predict(idata, data=new_data, kind='response') #predict the posterior predictive distribution

  #response parameters = sampling to find the uncertainty on the line (post. mean line)

# Make random draws from the posterior of the gravel size
y_pps_draws = idata.posterior_predictive["y"].values

# Reshape to (total_draws, x_points)
y_pps_draws = y_pps_draws.reshape(-1, len(xrange))

# Compute mean and HDI of the posterior predictive distribution
posterior_mean = y_pps_draws.mean(axis=0)
hdi_mean = az.hdi(idata.posterior["mu"].values.reshape(-1, len(xrange)), prob=0.95, axis=0)
hdi_pps = az.hdi(y_pps_draws, prob=0.95, axis=0)

# Plot both posterior for regression lines and the observations
plt.figure(figsize=(10, 6))
plt.plot(year_fw_filtered, log_pop_data, 'ok', label="Data")
plt.plot(xrange, posterior_mean, color="C0", label="Posterior mean")
plt.fill_between(
    xrange,
    hdi_mean[:, 0],
    hdi_mean[:, 1],
    alpha=0.3,
    color="C0",
    label="95% HDI average", #95% probability that the average gravel size will be in the HDI
    edgecolor = None
)

plt.fill_between(
    xrange,
    hdi_pps[:, 0],
    hdi_pps[:, 1],
    alpha=0.2,
    color="C0",
    label="95% HDI (predictive)", #95% probability that predicted gravel size (y given x) is in this HDI
    edgecolor = None
)

plt.legend()
plt.minorticks_on()
plt.xlabel('Year')
plt.ylabel('Log Fairy Wren Population')
plt.title('Predictive Regression Model for Fairy Wren ANBG Population')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
````

````python
## regular 95% HDI (Fairy Wren Log Population - Corrected)

xrange_hdi_regular = np.linspace(year_fw_filtered.min(), year_fw_filtered.max(), 1000) # Predict across a range of x values
new_data_hdi_regular = pd.DataFrame({"x": xrange_hdi_regular}) #dataframe with the new x-values
model.predict(idata, data=new_data_hdi_regular, kind='response_params') #predict the distribution of regression lines at each x-value

# Make random draws from the posterior of the regression lines
y_mean_draws = idata.posterior["mu"].values

# Reshape to (total_draws, x_points)
y_mean_draws = y_mean_draws.reshape(-1, len(xrange_hdi_regular))

# Compute posterior mean and HDI at each x point
posterior_mean = y_mean_draws.mean(axis=0)
hdi = az.hdi(y_mean_draws, prob=0.95, axis=0)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(year_fw_filtered, log_pop_data, 'ok', label="Observations")
plt.plot(xrange_hdi_regular, posterior_mean, color="C1", label="Posterior mean")
plt.fill_between(
    xrange_hdi_regular,
    hdi[:, 0],
    hdi[:, 1],
    alpha=0.3,
    color="C1",
    label="95% HDI (average)",
    edgecolor = None
)
plt.xlabel('Year') # Updated label
plt.ylabel('Log Population (Fairy Wren)') # Updated label
plt.title('Linear Regression Model for Log Fairy Wren Population with 95% HDI (Average)')
plt.legend()
plt.minorticks_on()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
````

````python
#fitting polynomial models

# Fit a straight-line model (first-order polynomial)
model_linear = bmb.Model("y ~ x", rdata)
idata_linear = model_linear.fit(idata_kwargs={"log_likelihood": True}, progressbar=False)

# Fit a quadratic model (second-order polynomial)
model_quad = bmb.Model("y ~ x + x2", rdata)
idata_quad = model_quad.fit(idata_kwargs={"log_likelihood": True}, progressbar=False)

# Fit a cubic model (third-order polynomial)
model_cubic = bmb.Model("y ~ x + x2 + x3", rdata)
idata_cubic= model_cubic.fit(idata_kwargs={"log_likelihood": True}, progressbar=False)

# Fit a quartic model (fourth-order polynomial)
model_quart = bmb.Model("y ~ x + x2 + x3 + x4", rdata)
idata_quart = model_quart.fit(idata_kwargs={"log_likelihood": True}, progressbar=False)
````

````python
#Compare polynomials and rank best suited for equation
comparison = az.compare({
    "linear": idata_linear,
    "quadratic": idata_quad,
    "cubic": idata_cubic,
    "quartic": idata_quart
})
print(comparison) #print the result
````

The most relevant regression model is found to be a *quadratic* polynomial, therefore will be utilised for analysis of this relationship.

The above plot represents the posterior mean of the quadratic regression model. This provides information to predict Fairy wren population size at the ANBG in respect to future years.

The ```95% HDI``` (Highest Density Interval) band represents the 95% probability envelope for the mean FW population size at a given year. This is considering prosterior plots for the ```intercept``` and gradient. 95% HDI is the 'credible range' for the regression line.

Now we can consider the ```sigma``` posterior distribution, representing deviations in population size away from the regression line. This will be our ```95% HDI (predictive)```.



````python
## Plotting with predictive prosteriors

# --- Definitions from cell 6a35531b (added to ensure variables are defined) ---
# Initialize the model with the quadratic formula using rdata
model_quadratic_final = bmb.Model("y ~ x + x2", rdata)
# Fit the model and get the inference data (idata) for the quadratic model
idata_quadratic_final = model_quadratic_final.fit(draws=2000, tune=2000, chains=10, progressbar=False, target_accept=0.95)
# --------------------------------------------------------------------------

plt.figure(figsize=(10, 6)) # Standardized figure size

# Predict across the range of filtered years
x_range_plot = np.linspace(year_fw_filtered.min(), year_fw_filtered.max(), 100)
new_data_plot = pd.DataFrame({"x": x_range_plot, "x2": x_range_plot**2}) # Include x2 for quadratic prediction

# Call predict to get posterior predictive samples
model_quadratic_final.predict(idata_quadratic_final, data=new_data_plot, kind='response')

# Make random draws from the posterior of the population size
y_pps_draws = idata_quadratic_final.posterior_predictive["y"].values

# Reshape to (total_draws, x_points)
y_pps_draws = y_pps_draws.reshape(-1, len(x_range_plot))

# Compute posterior mean and HDI at each x point
posterior_mean_quad = y_mean_draws_quad.mean(axis=0)
hdi_quad = az.hdi(y_mean_draws_quad, prob=0.95, axis=0) #regular/intercept prosterior hdi
hdi_pps = az.hdi(y_pps_draws, prob=0.95, axis=0) #predictive/sigma prosterior hdi


# Plot both posterior for regression lines and the observations
plt.plot(year_fw_filtered, average_pop_filtered, 'ok', label="Observations")
plt.plot(x_range_plot, posterior_mean_quad, color="C0", label="Posterior mean")
plt.fill_between(
    x_range_plot,
    hdi_quad[:, 0],
    hdi_quad[:, 1],
    alpha=0.3,
    color="C0",
    label="95% HDI (average)", #95% probability that the average population size will be in the HDI
    edgecolor = None
)

plt.fill_between(
    x_range_plot,
    hdi_pps[:, 0],
    hdi_pps[:, 1],
    alpha=0.2,
    color="C0",
    label="95% HDI (predictive)", #95% probability that predicted population size (y given x) is in this HDI
    edgecolor = None
)

plt.legend()
plt.minorticks_on()
plt.xlabel('Year')
plt.ylabel('Average Fairy wren Population')
plt.title('Predictive Regression Model for Fairy wren ANBG Population') # Updated title
plt.xlim(1993, 2021)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
````

The ```95% HDI (predictive)``` band represents the 95% probability envelope for the size of the population at a given year.
* A.k.a - at a random given year, the average population size of Fairy wrens at the ANBG should fall within the HDI predictive band 95% of the time.

````python
### trying cleaning the plot

# --- Definitions from cell 6a35531b (added to ensure variables are defined) ---
# Initialize the model with the quadratic formula using rdata
model_quadratic_final = bmb.Model("y ~ x + x2", rdata)
# Fit the model and get the inference data (idata) for the quadratic model
idata_quadratic_final = model_quadratic_final.fit(draws=2000, tune=2000, chains=10, progressbar=False, target_accept=0.95)
# --------------------------------------------------------------------------

# Plotting with predictive posteriors, including the Bird Index

plt.figure(figsize=(12, 7)) # Adjust figure size for better visibility

# Predict across the range of filtered years
x_range_plot = np.linspace(year_fw_filtered.min(), year_fw_filtered.max(), 100)
new_data_plot = pd.DataFrame({"x": x_range_plot, "x2": x_range_plot**2}) # Include x2 for quadratic prediction

# Call predict to get posterior predictive samples
model_quadratic_final.predict(idata_quadratic_final, data=new_data_plot, kind='response')

# Make random draws from the posterior of the population size
y_pps_draws = idata_quadratic_final.posterior_predictive["y"].values

# Reshape to (total_draws, x_points)
y_pps_draws = y_pps_draws.reshape(-1, len(x_range_plot))

# Compute posterior mean and HDI at each x point
posterior_mean_quad = y_mean_draws_quad.mean(axis=0)
hdi_quad = az.hdi(y_mean_draws_quad, prob=0.95, axis=0) #regular/intercept posterior hdi
hdi_pps = az.hdi(y_pps_draws, prob=0.95, axis=0) #predictive/sigma posterior hdi

# Plot both posterior for regression lines and the observations
plt.plot(year_fw_filtered, average_pop_filtered, 'k', linestyle='--', label="Population (Fairy Wren)") #-- line observations for fw data
plt.plot(x_range_plot, posterior_mean_quad, color="C0", label="Posterior mean (Fairy Wren)")
plt.fill_between(
    x_range_plot,
    hdi_pps[:, 0],
    hdi_pps[:, 1],
    alpha=0.2,
    color="C0",
    label="95% HDI (predictive Fairy Wren)",
    edgecolor = None
)

# Scale the Bird Index to match the starting value of the Fairy Wren population
# First value of bird_filtered is for 1993, first value of average_pop_filtered is for 1993
scaling_factor = average_pop_filtered[0] / bird_filtered[0]
scaled_bird_filtered = bird_filtered * scaling_factor

plt.plot(year_bird_filtered, scaled_bird_filtered, 'y', label='Scaled Bird Index (National)')

plt.legend()
plt.minorticks_on()
plt.xlabel('Year')
plt.ylabel('Population / Scaled Index') # Updated y-axis label for absolute comparison
plt.title('Predictive Regression Model for Fairy Wren Population vs. Scaled National Bird Index') # Updated title
plt.xlim(1993, 2021)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
````
